```python
# Advance Line plot
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.sin(x) * np.cos(x)

# Create the plot
plt.figure(figsize=(7, 4))

# Plot multiple lines
plt.plot(x, y1, label='sin(x)', color='blue', linestyle='-', linewidth=2, marker = "*")
plt.plot(x, y2, label='cos(x)', color='red', linestyle='--', linewidth=2)
plt.plot(x, y3, label='sin(x) * cos(x)', color='green', linestyle=':', linewidth=2)

# Add title and labels
plt.title('Advanced Line Plot Example', fontsize=16)
plt.xlabel('X-axis', fontsize=14)
plt.ylabel('Y-axis', fontsize=14)

# Add grid
plt.grid(True, linestyle='--', alpha=0.6)

# Add legend
plt.legend(loc='upper right', fontsize=12)

# Show the plot
plt.show()
```

```python
# Bar plot
import matplotlib.pyplot as plt
import numpy as np

# Sample data
categories = ['Category A', 'Category B', 'Category C', 'Category D']
values1 = [23, 45, 56, 78]
values2 = [34, 30, 45, 67]
values3 = [12, 25, 30, 40]

# Position of bars on the x-axis
x = np.arange(len(categories))

# Width of the bars
width = 0.2

# Create the plot
plt.figure(figsize=(10, 6))

# Plot grouped bars
plt.bar(x - width, values1, width, label='Group 1', color='skyblue', edgecolor='black')
plt.bar(x, values2, width, label='Group 2', color='lightgreen', edgecolor='black')
plt.bar(x + width, values3, width, label='Group 3', color='salmon', edgecolor='black')

# Add title and labels
plt.title('Advanced Bar Plot Example', fontsize=16)
plt.xlabel('Categories', fontsize=14)
plt.ylabel('Values', fontsize=14)

# Customize x-axis ticks
plt.xticks(x, categories, fontsize=12)

# Add grid
plt.grid(axis='y', linestyle='--', alpha=0.6)

# Add legend
plt.legend(loc='upper right', fontsize=12)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
# Histogram
import matplotlib.pyplot as plt
import numpy as np

data = np.random.randn(1000)

# Create the plot
plt.figure(figsize=(10, 6))

plt.hist(data, bins=20, color='green', alpha=0.7, edgecolor='black', linewidth=1, label = "Random Data")
plt.title('Customized Histogram', fontsize=14)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.6)

# Add legend
plt.legend(loc='upper right', fontsize=12)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
# Multiple Histograms on the Same Plot
import matplotlib.pyplot as plt
import numpy as np

data1 = np.random.normal(0, 1, 1000)  # Mean 0, Std 1
data2 = np.random.normal(3, 1, 1000)  # Mean 3, Std 1
# Create the plot
plt.figure(figsize=(10, 6))

plt.hist(data1, bins=30, color='blue', alpha=0.7, edgecolor='black', label='Dataset 1')
plt.hist(data2, bins=30, color='red', alpha=0.6, edgecolor='black', label='Dataset 2')

plt.title('Comparing Two Distributions')
plt.xlabel('Value')
plt.ylabel('Frequency')


plt.grid(axis='y', linestyle='--', alpha=0.6)

# Add legend
plt.legend(loc='upper right', fontsize=12)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
# Scatter Plot
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.random.rand(50)  # 50 random values for x
y = np.random.rand(50)  # 50 random values for y

# Create a basic scatter plot
plt.figure(figsize=(6, 4))
plt.scatter(x, y, color='green', marker='^', s=100)  # s = size of markers

# Add title and labels
plt.title('Basic Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Show the plot
plt.show()
```

```python
#Scatter Plot with Regression Line
from scipy.stats import linregress

# Sample data
x = np.random.rand(50)
y = 2 * x + np.random.normal(0, 0.1, 50)  # Linear relationship with noise

# Perform linear regression
slope, intercept, r_value, p_value, std_err = linregress(x, y)
line = slope * x + intercept  # Regression line

# Create a scatter plot with regression line
plt.figure(figsize=(6, 4))
plt.scatter(x, y, color='orange', label='Data Points')
plt.plot(x, line, color='blue', label='Regression Line')

# Add title, labels, and legend
plt.title('Scatter Plot with Regression Line')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()

# Show the plot
plt.show()
```

```python
# Pie Chart

import matplotlib.pyplot as plt

# Sample data
labels = ['Apples', 'Bananas', 'Cherries', 'Dates']
sizes = [15, 30, 45, 10]  # Sizes of each category (can be percentages or counts)

# Define custom colors
colors = ['gold', 'lightcoral', 'lightskyblue', 'lightgreen']
# Create a basic pie chart
plt.figure(figsize=(6, 6))
plt.pie(sizes, labels=labels, colors=colors ,autopct='%1.1f%%')

# Add title
plt.title('Basic Pie Chart')

# Show the plot
plt.show()
```
